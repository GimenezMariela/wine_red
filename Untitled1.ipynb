```python

"""
La matriz de confusión es una tabla que te dice exactamente:

¿Cuántos vinos malos fueron clasificados como malos?
¿Cuántos vinos bu fueron clasificados como malos (¡error!)?
¿Cuántos intermedios fueron clasificados como buenos? etc.
"""

def plot_confusion_matrix(y_test, y_prediction):
    '''Plotting Confusion Matrix'''
    cm = metrics.confusion_matrix(y_test, y_prediction) #compara dos listas. Crea una tabla (la matriz de confusión) que cuenta:
#cuántas veces el modelo acertó,
#cuántas veces se equivocó,
#y en qué se equivocó.
    ax = plt.subplot()
    ax = sns.heatmap(cm, annot=True, fmt='', cmap="Purples") #Esa gráfica usa colores para mostrar cuántos aciertos o errores hubo.

    ax.set_xlabel('Prediced labels', fontsize=18)
    ax.set_ylabel('True labels', fontsize=18)
    ax.set_title('Confusion Matrix', fontsize=25)
    #

    #pone palabras en vez de ver 0,1,2
    ax.xaxis.set_ticklabels(['Bad', 'Good', 'Middle'])
    ax.yaxis.set_ticklabels(['Bad', 'Good', 'Middle'])
    plt.show()
```

```python
"""
🧩 ¿Qué es este código?

Este código crea una gráfica que muestra dos cosas importantes sobre un modelo de predicción de vinos (o cualquier otra cosa que estés clasificando):

Cuántas veces el modelo acertó o se equivocó (Matriz de confusión)
Qué tan bien predice cada categoría (Reporte de clasificación)
Todo se ve en una sola imagen dividida en dos partes.

"""

def clf_plot(y_pred) :
    '''
    1) Ploting Confusion Matrix
    2) Plotting Classification Report

    '''
    # Esto compara:

    # y_test: lo que realmente era el vino (la verdad verdadera).
    # y_pred: lo que el modelo dijo que era.
    cm = metrics.confusion_matrix(y_test, y_pred)


    #👉 Esto crea otro informe (un poco más técnico) que mide:

    #Precisión (qué tan bien acierta).
    #Recall (qué tanto encuentra todos los casos).
    #F1-score (una mezcla de precisión y recall).
    cr = pd.DataFrame(metrics.classification_report(y_test, y_pred, digits=3, output_dict=True)).T

    #👉 Elimina una columna llamada "support" que no se necesita para esta gráfica
    cr.drop(columns='support', inplace=True)

    #Esto es para que:
    #En el panel izquierdo aparezca la matriz de confusión.
    #En el panel derecho el reporte de clasificación.
    fig, ax = plt.subplots(1, 2, figsize=(15, 5))

    # Left AX : Confusion Matrix
    ax[0] = sns.heatmap(cm, annot=True, fmt='', cmap="Purples", ax=ax[0]) #Cada cuadro dice cuántas veces ocurrió cada situación (aciertos y errores).
    ax[0].set_xlabel('Prediced labels', fontsize=18) # Abajo: lo que el modelo predijo.
    ax[0].set_ylabel('True labels', fontsize=18) #A la izquierda: lo que realmente era.
    ax[0].set_title('Confusion Matrix', fontsize=25)
    ax[0].xaxis.set_ticklabels(['Bad', 'Good', 'Middle'])
    ax[0].yaxis.set_ticklabels(['Bad', 'Good', 'Middle'])

    # Right AX : Classification Report

    #Las métricas: precisión, recall, F1-score.
    # Cada fila es una clase ("bueno", "malo", etc.)
    ax[1] = sns.heatmap(cr, cmap='Purples', annot=True, linecolor='white', linewidths=0.5, ax=ax[1])

    ax[1].xaxis.tick_top()
    ax[1].set_title('Classification Report', fontsize=25)
    plt.show()

    """
    📊 Resultado visual esperado:

  Verás dos cuadros uno al lado del otro:

  🔹 Izquierda (Matriz de Confusión):
  Cuántas veces acertó el modelo y cuántas veces se equivocó, y cómo se equivocó.

  🔹 Derecha (Reporte de Clasificación):
  Qué tan bueno fue el modelo en cada categoría."""
```
