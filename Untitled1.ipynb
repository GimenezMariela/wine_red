```python

"""
La matriz de confusin es una tabla que te dice exactamente:

Cuntos vinos malos fueron clasificados como malos?
Cuntos vinos bu fueron clasificados como malos (error!)?
Cuntos intermedios fueron clasificados como buenos? etc.
"""

def plot_confusion_matrix(y_test, y_prediction):
    '''Plotting Confusion Matrix'''
    cm = metrics.confusion_matrix(y_test, y_prediction) #compara dos listas. Crea una tabla (la matriz de confusin) que cuenta:
#cuntas veces el modelo acert,
#cuntas veces se equivoc,
#y en qu se equivoc.
    ax = plt.subplot()
    ax = sns.heatmap(cm, annot=True, fmt='', cmap="Purples") #Esa grfica usa colores para mostrar cuntos aciertos o errores hubo.

    ax.set_xlabel('Prediced labels', fontsize=18)
    ax.set_ylabel('True labels', fontsize=18)
    ax.set_title('Confusion Matrix', fontsize=25)
    #

    #pone palabras en vez de ver 0,1,2
    ax.xaxis.set_ticklabels(['Bad', 'Good', 'Middle'])
    ax.yaxis.set_ticklabels(['Bad', 'Good', 'Middle'])
    plt.show()
```

```python
"""
 Qu es este cdigo?

Este cdigo crea una grfica que muestra dos cosas importantes sobre un modelo de prediccin de vinos (o cualquier otra cosa que ests clasificando):

Cuntas veces el modelo acert o se equivoc (Matriz de confusin)
Qu tan bien predice cada categora (Reporte de clasificacin)
Todo se ve en una sola imagen dividida en dos partes.

"""

def clf_plot(y_pred) :
    '''
    1) Ploting Confusion Matrix
    2) Plotting Classification Report

    '''
    # Esto compara:

    # y_test: lo que realmente era el vino (la verdad verdadera).
    # y_pred: lo que el modelo dijo que era.
    cm = metrics.confusion_matrix(y_test, y_pred)


    # Esto crea otro informe (un poco ms tcnico) que mide:

    #Precisin (qu tan bien acierta).
    #Recall (qu tanto encuentra todos los casos).
    #F1-score (una mezcla de precisin y recall).
    cr = pd.DataFrame(metrics.classification_report(y_test, y_pred, digits=3, output_dict=True)).T

    # Elimina una columna llamada "support" que no se necesita para esta grfica
    cr.drop(columns='support', inplace=True)

    #Esto es para que:
    #En el panel izquierdo aparezca la matriz de confusin.
    #En el panel derecho el reporte de clasificacin.
    fig, ax = plt.subplots(1, 2, figsize=(15, 5))

    # Left AX : Confusion Matrix
    ax[0] = sns.heatmap(cm, annot=True, fmt='', cmap="Purples", ax=ax[0]) #Cada cuadro dice cuntas veces ocurri cada situacin (aciertos y errores).
    ax[0].set_xlabel('Prediced labels', fontsize=18) # Abajo: lo que el modelo predijo.
    ax[0].set_ylabel('True labels', fontsize=18) #A la izquierda: lo que realmente era.
    ax[0].set_title('Confusion Matrix', fontsize=25)
    ax[0].xaxis.set_ticklabels(['Bad', 'Good', 'Middle'])
    ax[0].yaxis.set_ticklabels(['Bad', 'Good', 'Middle'])

    # Right AX : Classification Report

    #Las mtricas: precisin, recall, F1-score.
    # Cada fila es una clase ("bueno", "malo", etc.)
    ax[1] = sns.heatmap(cr, cmap='Purples', annot=True, linecolor='white', linewidths=0.5, ax=ax[1])

    ax[1].xaxis.tick_top()
    ax[1].set_title('Classification Report', fontsize=25)
    plt.show()

    """
     Resultado visual esperado:

  Vers dos cuadros uno al lado del otro:

   Izquierda (Matriz de Confusin):
  Cuntas veces acert el modelo y cuntas veces se equivoc, y cmo se equivoc.

   Derecha (Reporte de Clasificacin):
  Qu tan bueno fue el modelo en cada categora."""
```
